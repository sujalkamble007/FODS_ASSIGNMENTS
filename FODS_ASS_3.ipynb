```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
df=pd.read_csv("C:/Users/SUJAL/Desktop/dataset/iris - iris.csv")
```

```python
df.head(10)
```

```python
columns=df.iloc[:,[1,2,3,4]]
columns
```

```python
correlation_matrix = columns.corr()
print(correlation_matrix)
```

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
plt.figure(figsize=(2,9))
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
```

The correlation matrix of the Iris dataset reveals the relationships between its features: sepal length, sepal width, petal length, and petal width.

Strong positive correlations, such as between petal length and petal width, indicate that these two features increase together and may carry similar information.

 Conversely, weaker correlations, like between sepal width and the other features, suggest that sepal width provides unique insights. This analysis helps identify redundant features (e.g., petal length and petal width) and better understand the data structure for tasks like feature selection and model building.
