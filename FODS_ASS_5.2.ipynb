```python
import numpy as np
import pandas as pd
```

```python
air_quality=pd.read_csv("Air_Quality.csv")
air_quality
```

```python
air_quality.info()
```

```python
air_quality.head(10)
```

```python
#1.Data Cleaning:
#printing the null values
null_values=air_quality.isnull().sum()
print(null_values)
```

```python
# Convert 'Message' column to object type
air_quality['Message'] = air_quality['Message'].astype('object')

# Fill missing values with "Message is not given"
air_quality['Message'].fillna("Message is not given", inplace=True)
```

```python
air_quality.head(10)
```

```python
null_values=air_quality.isnull().sum()
print(null_values)
#all null values has been removed.
```

```python
#giving date time format for "Start_Date" Column
air_quality['Start_Date'] = pd.to_datetime(air_quality['Start_Date'], errors='coerce')
air_quality.head(10)
```

```python
# remove duplicate value
air_quality.drop_duplicates(inplace=True)
air_quality.head(5)
```

```python
#2.Data Transformation
# one-hot encoding(categorical into numerical format )
pd.get_dummies(air_quality, columns=['Geo Type Name', 'Geo Place Name','Name'], drop_first=True)
```

```python
# Extract the year from 'Start_Date'
air_quality['Start_Date'].dt.year
```
