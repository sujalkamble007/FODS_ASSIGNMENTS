```python
sns.boxplot(heart_dis["Cholesterol"])
#Boxplot of Cholesterol
```

```python
#Define a function to remove outliers based on IQR
def remove_outliers_from_boxplot(heart_dis, x):
    Q1 = heart_dis["Cholesterol"].quantile(0.25)
    Q3 = heart_dis["Cholesterol"].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return heart_dis[(heart_dis["Cholesterol"] >= lower_bound) & (heart_dis["Cholesterol"] <= upper_bound)]

# Apply outlier removal on 'Cholesterol'
heart_dis = remove_outliers_from_boxplot(heart_dis, 'Cholesterol')
```

```python
sns.boxplot(heart_dis["Cholesterol"])
#outliers has been removed from Cholesterol
```
